# forecasting 
Let's take a look at what would happen if demand(sales) and inventory increase by 10%


```python
mergeddf['inflow_outflow_ten'] = mergeddf['prod_fcst'].fillna(0) - (mergeddf['demand'].fillna(0)*1.1)
mergeddf['starting_inventory_ten'] = mergeddf['starting_inventory']
mergeddf.loc[mergeddf['period_id_mnth'] > '2022-03-01', 'starting_inventory_ten'] = np.nan

mergeddf['starting_inventory_ten'] = mergeddf['starting_inventory_ten'].fillna(mergeddf['inflow_outflow_ten'])
mergeddf.loc[mergeddf['period_id_mnth'] < '2022-03-01', 'starting_inventory_ten'] = np.nan
mergeddf['inventory_forecast_ten'] = mergeddf.groupby(['item'])['starting_inventory_ten'].cumsum()
```

```python
simdemand = mergeddf.loc[mergeddf['item'] == 'Item 06129'].copy()
simdemand['demand_ten'] = simdemand['demand'] * 1.1
simdemand.loc[mergeddf['period_id_mnth'] <= '2022-03-01', 'demand_ten'] = np.nan
simdemand = simdemand[['period_id_mnth','inventory', 'inventory_forecast','inventory_forecast_ten','demand', 'demand_ten', 'prod_fcst']]
simdemand['period_id_mnth'] = pd.to_datetime(simdemand['period_id_mnth'] )
simdemand = simdemand.set_index('period_id_mnth')
plt.figure()

simdemand.plot(figsize=(15, 6))
```

```python
simdemand.to_csv('simdemand.csv',index=False)
```

## forecasting 
Now let's take a look at a different scale, what would it look like if there were no new production to keep up with a demand(sales) that was increasing by 10%
You can see that when the inventory runs out in March and with no new production, the inventory falls into the negative, and even when production resumes the yellow line of our forcast levels out, but can't catch up with the increased demand. 

```python
sim_prod = mergeddf.loc[mergeddf['item'] == 'Item 00048'].copy()
sim_prod['demand_ten'] = sim_prod['demand'] * 1.1
sim_prod.loc[mergeddf['period_id_mnth'] <= '2022-03-01', 'demand_ten'] = np.nan
sim_prod = sim_prod[['period_id_mnth','inventory', 'inventory_forecast','inventory_forecast_ten','demand',  'prod_fcst']]
sim_prod['period_id_mnth'] = pd.to_datetime(sim_prod['period_id_mnth'] )
sim_prod = sim_prod.set_index('period_id_mnth')
plt.figure()

sim_prod.plot(figsize=(15, 6))
```
